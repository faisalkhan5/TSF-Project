### **TASK 6 : Classify Different Species of flowers in Iris Dataset**

Step 1 : Importing libraries in Python

```python
import sklearn.datasets as datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Step 2 : Loading the iris dataset and Forming a dataframe


```python
iris=datasets.load_iris()

df=pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.head(10))
```

```python
y=iris.target
print(y)
```

## Train Test Split

Step 3 : Let's split up the data into a training set and a test set!

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.20)
```

### Now let us define the Decision Tree Algorithm

```python
# Defining the decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

## Prediction and Evaluation 

Step 4 : Let's evaluate our decision tree.


```python
predictions = dtree.predict(X_test)

from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
